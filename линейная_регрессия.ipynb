```python
# y = b1*x + b0 линейная регрессия

import numpy as np # import library numpy
from sklearn.linear_model import LinearRegression # import class LinearRegression
# input date
x = np.array([2508, 2572, 2408, 2522, 2700, 2531, 2390]).reshape((-1, 1))
y = np.array([2406, 2464, 2336, 2281, 2641, 2385, 2297])
model = LinearRegression().fit(x, y)
print(f"intercept: {model.intercept_}") # b0
print(f"coefficients: {model.coef_}") #b1

# построение графиков для линейной регрессии
import matplotlib.pyplot as plt
plt.plot(x, y, 'ro')

x1 = np.linspace(2000, 2800, 5)
y1 = 1.03083736*x1 -194.95622280320413
plt.plot(x1, y1)



```
